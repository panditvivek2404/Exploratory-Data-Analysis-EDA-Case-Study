* Detect outliers , missing values or errors or duplicates

* Uncover relationship between variables

* Generate hypothesis and guides further analysis

# What to do with these 52 values

* Drop them ?

* Replace them ?

* When to drop the null values = If the count of null values is less than 30 % of your data

* When to replace the null values = If the count of the null values is more than 30 % of the data

```python
1338 * 30/100
```

```python
# drop the null values as it is less than 30 % of the data

data.dropna(inplace = True)
```

```python
print(data.isnull().sum().sum())
```

```python
data.shape
```

* If we will select the replace option ? Then by which we are going to replace the missing value

* If the data type of the column is object = We will replace the missing value with the mode of that column

* If the datatype of column is numerical = Replace the missing value with the mean / median of that column

```python
for i in data.columns:
  print(i)
```

```python
for i in data.columns:
  if data[i].dtype == 'object':
    data[i] = data[i].fillna(data[i].mode()[0])
  else:
    data[i] = data[i].fillna(data[i].mean())
```

```python
data['smoker'].mode()
```

```python
data['smoker'].value_counts()
```

```python
# How to find the dupliacte records in our data

print(data.duplicated().sum())
```

```python
# How to delete the duplicate records in our data

data.drop_duplicates(inplace = True)
```

```python
sns.boxplot(y = data['age'])
```

```python
# Find outliers using boxplot

for i in data.columns:
  # Draw a boxplot if the datatype of the column is numerical '
  if data[i].dtype == 'int64' or data[i].dtype == 'float64':
    plt.boxplot(data[i])
    plt.xlabel(i)
    plt.show()
```

```python
# How to remove the records that are outliers

for i in data.columns:
  if data[i].dtype == 'int64' or data[i].dtype == 'float64':
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    upper_fence = Q3 + 1.5 * IQR
    lower_fence = Q1 - 1.5 * IQR
    data = data[(data[i] >= lower_fence) & (data[i] <= upper_fence)]
```

```python
# Find outliers using boxplot

for i in data.columns:
  # Draw a boxplot if the datatype of the column is numerical '
  if data[i].dtype == 'int64' or data[i].dtype == 'float64':
    plt.boxplot(data[i])
    plt.xlabel(i)
    plt.show()
```
